## DATA VISUALIZATION

```python
## generating histograms for numerical variables
df.hist(figsize=(15,15),color="indigo")
```

```python
## histogram for numerical values
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], kde=True,color="red")
plt.title("Histogram of AGE")
plt.show()
```

```python
## bar charts for categorical variables
plt.figure(figsize=(4,4))
sns.countplot(x="LUNG_CANCER",data=df,color="violet")
```

```python
## correlation matrix
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
```

From the EDA operation, we can conclude the following:

* The dataset has no missing values.
* The age distribution is skewed to the right, indicating that most patients are older.
* The majority of patients are non-smokers.
* The FEV1 values are normally distributed, but with some outliers.
* There is a strong positive correlation between FEV1 and FVC.
* The correlation matrix shows that there are many strong correlations between the features, which may indicate multicollinearity.
