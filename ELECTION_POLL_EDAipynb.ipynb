# **NUMERICAL DATA PLOTTING**

```python
## making a histogram from the numeric values columns
plt.figure(figsize=(10,8))
sns.histplot(df["age"],kde=True,color="green")
## title of the histogram
plt.title("Histogram of AGE")
plt.show()
```

```python
## making histogram for weight column
plt.figure(figsize=(10,8))
sns.histplot(df["weight1"],kde=True,color="black")
plt.title("Histogram of Weight")
plt.show()
```

# **CATEGORICAL DATA PLOTTING**

```python
## bar chart for sex column
plt.figure(figsize=(10, 8))
sns.countplot(x='sex', data=df,color="brown")
plt.title("Bar Chart of SEX")
plt.show()
```

```python
## bar chat for federal district column
plt.figure(figsize=(10, 6))
sns.countplot(x='federal_district', data=df,color="yellow")
plt.title("Bar Chart of FEDERAL DISTRICT")
plt.show()
```

```python
## bar chat fot type_of_city column
plt.figure(figsize=(10, 8))
sns.countplot(x='type_of_city', data=df,color="pink")
plt.title("Bar Chart of type of city")
plt.show()
```

```python
## bar chart for will_vote column
plt.figure(figsize=(10, 8))
sns.countplot(x='will_vote', data=df,color="blue")
plt.title("Bar Chart of will vote")
plt.show()
```

```python
## bar chart for education column
plt.figure(figsize=(10, 8))
sns.countplot(x='education', data=df,color="red")
plt.title("Bar Chart of education")
plt.show()
```

```python
## correlation matrix
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
```

# **PAIRPLOT FOR NUMERIC DATA**

```python
sns.pairplot(df[['age', 'weight1']])
plt.show()
```

```python
corr_matrix = pd.crosstab(df['sex'], df['federal_district']).apply(lambda r: r/r.sum())
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title("Correlation Matrix for SEX and FEDERAL DISTRICT")
plt.show()
```

# **FINDING OUTLIERS**

```python
## outliers in age column
plt.figure(figsize=(10, 6))
sns.boxplot(x='age', data=df)
plt.title("Boxplot of AGE")
plt.show()

```

```python
## finding outliersin weight column
plt.figure(figsize=(10, 6))
sns.boxplot(x='weight1', data=df,color="")
plt.title("Boxplot of WEIGHT")
plt.show()
```
